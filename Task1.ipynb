```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

Import the dataset.

```python
data = pd.read_csv("data_Q1.csv")
```

View the first 5 entries.

```python
data.head()
```

### As a database comprehension task:

```python
def get_species(SL, SW, PL, PW):
    d1 = data.loc[data.SepalLengthCm==SL]
    d2 = d1.loc[d1.SepalWidthCm==SW]
    d3 = d2.loc[d2.PetalLengthCm==PL]
    d4 = d3.loc[d3.PetalWidthCm==PW]
    return d4.Species
```

### As a multiclass classification task:

```python
input_data = data.drop(columns='Id')
```

Visualise the pairwise relationships between features for each class.

```python
sns.pairplot(input_data, hue='Species')
plt.show()
```

Separate targets and features, then perform a 70/30 train-test split.

```python
x = input_data.drop(columns='Species')
y = input_data.Species
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, train_size=0.7)
```

Train a linear SVM to perform classification.

```python
classifier = svm.LinearSVC(max_iter=5000).fit(train_x, train_y)
```

Make a prediction.

```python
pred_y = classifier.predict(test_x)
```

Visualise the output in a confusion matrix.

```python
cm = confusion_matrix(test_y, pred_y)
ConfusionMatrixDisplay(cm).plot()
plt.show()
```
